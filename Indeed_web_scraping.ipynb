# Webscraping of uk.indeed.com using Beautiful Soup

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

# Function to extract pagewise job posts

## Search criteria : Position: Data Scientist and Location: London

```python
def extract(page):
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.111 Safari/537.36'}
    url = f'https://uk.indeed.com/jobs?q=data+scientist&l=London%2C+Greater+London&start={page}'
    r=requests.get(url,headers)
    soup=BeautifulSoup(r.content,"html.parser")
    return soup
```
