```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# define function extract pagewise

def extract(page):
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.111 Safari/537.36'}
    url = f'https://uk.indeed.com/jobs?q=data+scientist&l=London%2C+Greater+London&start={page}'
    r=requests.get(url,headers)
    soup=BeautifulSoup(r.content,"html.parser")
    return soup
```

```python
# define function transform to capture required details

def transform(soup):
    error_=[]
    divs=soup.findAll('div',class_='job_seen_beacon')
    for item in divs:
        title=item.find('span').text # title is not quite right 
        #print(title)
    
        company_name=item.find('span',class_="companyName").text
        #print(company_name)
    
 
        location=item.find('div',class_='companyLocation').text
        #print(location)
        
        try:
            salary=item.find('div',attrs={'class':'salary-snippet'})
            #print(salary)
        except:
            salary=''
            #print(salary)        
        
        summary=item.find('div',class_='job-snippet').text.strip().replace('\n','')
        #print(summary)
        job={
            'title':title,
            'company_name':company_name,
            'location':location,
            'salary':salary,
            'summary':summary
        }
        try:
            job_list.append(job)
        except:
            error_.append(company_name)
    return 
```

```python
job_list=[]

# extracting 3 pages of job details
for i in range(0,40,10):
    print(f'page {i}')
    c=extract(i)
    transform(c)
    #print(job_list)
    
# inserting the job_list dictionary to a dataframe    
df=pd.DataFrame(job_list)
```
